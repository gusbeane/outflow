```python
import numpy as np
import matplotlib.pyplot as plt
import arepo
import matplotlib as mpl
```

```python
basepath = '/n/holylfs05/LABS/hernquist_lab/Users/abeane/outflow/'
```

```python
sn = arepo.Snapshot(basepath + 'runs/GALAKOS_fg0.1/lvl4/output', 200, combineFiles=True)
```

```python
nres = 256
rng = [[-10, 10], [-10, 10]]
vmin = 5E-4
vmax = 1E-1

dx = (rng[0][1]-rng[0][0]) / nres
dy = (rng[1][1]-rng[1][0]) / nres
surf_area = dx * dy

center = np.array([sn.BoxSize, sn.BoxSize, sn.BoxSize])/2.
pos = sn.part2.pos.value - center
mass = np.full(sn.NumPart_Total[2], sn.MassTable[2])

Hxy_s, _, _ = np.histogram2d(pos[:,0], pos[:,1], bins=(nres, nres), range=rng, weights=mass/surf_area)
```

```python
fig, ax = plt.subplots(1, 1)
ax.imshow(Hxy_s.T, origin='lower', norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
```
