```python
import numpy as np
import arepo
import matplotlib.pyplot as plt
```

```python
basepath = '/n/holylfs05/LABS/hernquist_lab/Users/abeane/outflow/'
```

```python
def bin_sfr(t, sfr, bin_width=0.01):
    bin_left = 0.0
    bin_right = bin_left + bin_width
    
    ave_sfr = []
    ave_t = []
    
    max_val = np.max(t)
    while bin_left < max_val:
        key = np.logical_and(t >=bin_left, t<bin_right)
        ave_sfr.append(np.mean(sfr[key]))
        ave_t.append(np.mean(t[key]))
        
        bin_left += bin_width
        bin_right += bin_width
        
    return np.array(ave_t), np.array(ave_sfr)
    
```

```python
sfh_file = np.genfromtxt(basepath+'runs/GALAKOS_fg0.1/lvl4/output/sfr.txt')
time, sfr = bin_sfr(sfh_file[:,0], sfh_file[:,2])
plt.plot(time, sfr, label='GALAKOS lvl4')

sfh_file = np.genfromtxt(basepath+'runs/GALAKOS_fg0.1/lvl3/output/sfr.txt')
time, sfr = bin_sfr(sfh_file[:,0], sfh_file[:,2])
plt.plot(time, sfr, label='GALAKOS lvl3')

sfh_file = np.genfromtxt(basepath+'runs/GALAKOS_fg0.1/lvl2/output/sfr.txt')
time, sfr = bin_sfr(sfh_file[:,0], sfh_file[:,2])
plt.plot(time, sfr, label='GALAKOS lvl2')

plt.legend(frameon=False)
plt.xlabel('t [Gyr]')
plt.ylabel('SFR [Msun/yr]')
plt.ylim(0, None)
plt.xlim(0, None)

# plt.axhline()
```
